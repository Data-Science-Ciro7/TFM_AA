# ML SUBSAMPLE SYNTHETIC RV CURVES EXAMPLES (FINAL)

In this notebook we just show two examples of the synthetic RV curves generated from the statistical characteristics of the ML subsample: one for a pulsating star and another one for a non-pulsating star.

## Modules and configuration

### Modules

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white", {'figure.figsize':(15,10)})
```

### Configuration

```python
# CONFIGURATION:
IMAGE_FOLDER = "./img/"

DATASETS_FOLDER = "../data/SYNTH_RV_SAMPLES/"
DATASETS_SUMMARY_FILE = "../data/SYNTH_RV_SAMPLES/RV_FINAL_ML_SyntheticDatasets.csv"
DS1_FOLDER = "../data/SYNTH_RV_SAMPLES/S1_FINAL_ts_files/"
DS2_FOLDER = "../data/SYNTH_RV_SAMPLES/S2_FINAL_ts_files/"
DS3_FOLDER = "../data/SYNTH_RV_SAMPLES/S3_FINAL_ts_files/"
DS4_FOLDER = "../data/SYNTH_RV_SAMPLES/S4_FINAL_ts_files/"
```

### Functions

```python
def draw_curves(times, values, labels, value_label, title, fig_filename=None):
    '''Draws several time series curves in the same plot'''
    # Plots and saves the figure
    #kwargs = dict({'ms': 2.0})
    plt.figure(figsize=(10,7))
    plt.title(title, fontsize=16)
    #plt.grid(axis='both', alpha=0.75)
    plt.xlabel("Time [BJD]", fontsize=12)
    plt.ylabel(value_label, fontsize=12)
    for i in range(0, len(times)):
        g = sns.scatterplot(x=times[i], y=values[i], label=labels[i])
    plt.legend();
    ylabels = ['{:,.2f}'.format(y) for y in g.get_yticks()]
    g.set_yticklabels(ylabels)
    # Save the image:
    if fig_filename is None:
        pass
    else:
        plt.savefig(fig_filename, format='jpg', bbox_inches='tight')
```

## Load the synthetic datasets table

```python
ds = pd.read_csv(DATASETS_SUMMARY_FILE, sep=',', decimal='.')
ds.head()
```

## Plot examples

### Pulsating star

We take one of the pulsating stars, for example:

```python
# TRIAL AND ERROR - IN ORDER TO CHOSE A CLEAR EXAMPLE THAT IS SHOWN WELL IN THE PLOT:
CHOSEN_INDEX = 1
```

```python
p_id = ds.loc[ds['Pulsating']==True, 'ID'].reset_index(drop=True)[CHOSEN_INDEX]
p_id
```

Star properties:

```python
p_frequency = ds.loc[ds['Pulsating']==True, 'frequency'].reset_index(drop=True)[CHOSEN_INDEX]
p_amplitude = ds.loc[ds['Pulsating']==True, 'amplitudeRV'].reset_index(drop=True)[CHOSEN_INDEX]
p_phase = ds.loc[ds['Pulsating']==True, 'phase'].reset_index(drop=True)[CHOSEN_INDEX]
p_S2_error_mean = ds.loc[ds['Pulsating']==True, 'S2_errorRV_mean'].reset_index(drop=True)[CHOSEN_INDEX]
p_S2_error_stdev = ds.loc[ds['Pulsating']==True, 'S2_errorRV_stdev'].reset_index(drop=True)[CHOSEN_INDEX]
p_S4_error_mean = ds.loc[ds['Pulsating']==True, 'S4_errorRV_mean'].reset_index(drop=True)[CHOSEN_INDEX]
p_S4_error_stdev = ds.loc[ds['Pulsating']==True, 'S4_errorRV_stdev'].reset_index(drop=True)[CHOSEN_INDEX]
```

```python
print("Noise distribution index: %d" \
      %ds.loc[ds['Pulsating']==True, 'S2_errorRV_dist_idx'].reset_index(drop=True)[CHOSEN_INDEX])
print("Noise distribution type: %s" \
      %ds.loc[ds['Pulsating']==True, 'S2_errorRV_dist_name'].reset_index(drop=True)[CHOSEN_INDEX])
print("")
print("Sampling index: %d" \
      %ds.loc[ds['Pulsating']==True, 'S3_sampling_idx'].reset_index(drop=True)[CHOSEN_INDEX])
```

#### S1 and S2

```python
ds1_file = ds.loc[ds['Pulsating']==True, 'S1_file'].reset_index(drop=True)[CHOSEN_INDEX]
ds1_file
```

```python
ds1_curve = pd.read_csv(ds1_file, sep=' ', decimal='.',
                        names=['time', 'value', 'error'])
ds1_time = ds1_curve['time'] 
ds1_value = ds1_curve['value']
ds1_curve.head()
```

```python
ds2_file = ds.loc[ds['Pulsating']==True, 'S2_file'].reset_index(drop=True)[CHOSEN_INDEX]
ds2_file
```

```python
ds2_curve = pd.read_csv(ds2_file, sep=' ', decimal='.',
                        names=['time', 'value', 'error'])
ds2_time = ds2_curve['time'] 
ds2_value = ds2_curve['value']
ds2_curve.head()
```

#### Draw S1 and S2

```python
draw_curves(times=[ds2_time, ds1_time], values=[ds2_value, ds1_value],
            labels=["S2 - With noise", "S1 - Noiseless"],
            value_label="Radial velocity [$m\;s^{-1}$]",
            title="Synthetic curves S1 and S2 for pulsating object " + p_id + "\n" + \
                "Star pulsation parameters: $\\nu$: %.2f $d^{-1}$, A: %.2f $ms^{-1}$, $\delta$: %.2f.\n" %(p_frequency, p_amplitude, p_phase) + \
                "S2 error parameters: mean: %.2f $ms^{-1}$, stdev: %.2f $ms^{-1}$" %(p_S2_error_mean, p_S2_error_stdev),
            fig_filename=IMAGE_FOLDER + "SD_29_01_Pulsating_S1_S2.jpg")
```

We see that there is too many points to discern the undelying pulsation, and notice also that there is a strange effect in the right end of the plot (we are not using transparency), so we need to zoom-in.

#### Zoom in the time baseline

```python
ZOOM_N_POINTS = 1000 # To only show a reduced number of periods (i.e. not 100 days), to reduce clutter
TIME_BASELINE = ZOOM_N_POINTS * 100.0 / len(ds1_time)
draw_curves(times=[ds2_time[:ZOOM_N_POINTS], ds1_time[:ZOOM_N_POINTS]],
            values=[ds2_value[:ZOOM_N_POINTS], ds1_value[:ZOOM_N_POINTS]],
            labels=["S2 - With noise", "S1 - Noiseless"],
            value_label="Radial velocity [$m\;s^{-1}$]",
            title="Synthetic curves S1 and S2 for pulsating object " + p_id + \
                " (zoom in %d points = $\sim%.2f\;d$) \n" %(ZOOM_N_POINTS, TIME_BASELINE) + \
                "Star pulsation parameters: $\\nu$: %.2f $d^{-1}$, A: %.2f $ms^{-1}$, $\delta$: %.2f.\n" %(p_frequency, p_amplitude, p_phase) + \
                "S2 error parameters: mean: %.2f $ms^{-1}$, stdev: %.2f $ms^{-1}$" %(p_S2_error_mean, p_S2_error_stdev),
            fig_filename=IMAGE_FOLDER + "SD_29_01B_Pulsating_S1_S2_Zoom.jpg")
```
